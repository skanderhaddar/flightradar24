```python
from kafka import KafkaProducer
import json

# Configuration du producteur Kafka
conf = {'bootstrap.servers': "localhost:9092"}

# Création de l'instance du producteur Kafka
producer = KafkaProducer(bootstrap_servers="localhost:9092")

# Fonction de livraison du message
def delivery_callback(err, msg):
    if err:
        print("Erreur lors de l'envoi - {}".format(err))
    else:
        print("Message envoyé : {}".format(msg.value().decode('utf-8')))

# Exemple de données au format JSON
data = {"nom": "haddar" , "prenom":"skander"}

# Envoi du message JSON à un sujet Kafka
producer.send('skander_info', json.dumps(data).encode('utf-8')).add_callback(delivery_callback)

# Attendre que le message soit envoyé
producer.flush()
```

```python
from kafka import KafkaConsumer
import json

# Configuration du consommateur Kafka
conf = {'bootstrap.servers': "localhost:9092", 'group.id': "my_consumer_group"}

# Création de l'instance du consommateur Kafka
consumer = KafkaConsumer('skander_info', 
                         bootstrap_servers='localhost:9092',
                         auto_offset_reset='earliest',
                         enable_auto_commit=True,
                         group_id='my_consumer_group',
                         value_deserializer=lambda x: json.loads(x.decode('utf-8')))

# Consommation de messages
k=0
for message in consumer:
    print(k)
    k = k+1
    message_value = message.value
    print("Message reçu:", message_value)
    #break
# Fermer le consommateur
consumer.close()
```
